# LC Oscillator

image.png

How to define your circuit?  
 = ["device", "node_1", "node_2", "value"]

```python
# Circuit Text Definition:

LC_Circuit_Text = """#LC_Circuit
branches:
- ["C",1,0,Ec = 1]
- ["L",1,0,El = 1]
 """
```

```python
# Circuit definition

LC_Circuit = scq.Circuit(LC_Circuit_Text, from_file=False, ext_basis='harmonic')
```

Lets find the Lagrangian:

image.png

```python
# Calculate the system lagrangian:

LC_Circuit.sym_lagrangian()
```

Lets find the Hamiltonian:

image.png

```python
# Calculate the system Hamiltonian:

LC_Circuit.sym_hamiltonian()
```

```python
# Plot the potential
theta_1 = np.linspace(-2*np.pi, 2*np.pi,100)
LC_Circuit.plot_potential(θ1=np.linspace(-np.pi, np.pi))
plt.grid()
```

```python
# Calculate the eigenvalues:
LC_Circuit.eigenvals()
```

```python
LC_Circuit.eigenvals()[1]-LC_Circuit.eigenvals()[0]
```

```python
LC_Circuit.eigenvals()[2]-LC_Circuit.eigenvals()[1]
```

# Josephson Circuit

image.png

```python
# Circuit Text Definition:

CJJ_Circuit_Text = """#CJJ_Circuit
branches:
- ["JJ",1,0,EJ = 10, ECJ = 10]
- ["C",1,0, EC = 10]
"""
```

```python
# Circuit definition

CJJ_Circuit = scq.Circuit(CJJ_Circuit_Text, from_file=False)
```

Lets find the Lagrangian:

image.png

```python
# Calculate the system lagrangian:

CJJ_Circuit.sym_lagrangian()
```

Lets find the Hamiltonian:

image.png

```python
# Calculate the system Hamiltonian:
# ng1 correspond to Offset charges

CJJ_Circuit.sym_hamiltonian()
```

image.png

```python
CJJ_Circuit.plot_potential(θ1=np.linspace(-np.pi, np.pi))
plt.grid()
```

```python
CJJ_Circuit.eigenvals()[1]-CJJ_Circuit.eigenvals()[0]
```

```python
CJJ_Circuit.eigenvals()[2]-CJJ_Circuit.eigenvals()[1]
```

Lets compare the potentials from LC & CJJ Circuits

```python
CJJ_Circuit.potential_symbolic
```

```python
LC_Circuit.potential_symbolic
```

```python
theta_1 = np.linspace(-np.pi, np.pi)
potential_CJJ = -10*np.cos(theta_1)
potential_CL = 0.5*2*theta_1*theta_1

plt.plot(theta_1, potential_CJJ, label = "C+JJ Potential")
plt.plot(theta_1, potential_CL, label = "C+L Potential")
plt.legend()
plt.grid()
```
